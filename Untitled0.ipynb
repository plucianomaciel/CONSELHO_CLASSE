```python
# Salva o script como arquivo .py
codigo = '''
import pandas as pd
import tkinter as tk
from tkinter import filedialog

# Abas de interesse
abas_turmas = [
    '6A', '6B', '6C',
    '7A', '7B', '7C',
    '8A', '8B', '8C',
    '9A', '9B',
    '1A', '1B',
    '2A', '3A'
]

# Janela para selecionar os arquivos
root = tk.Tk()
root.withdraw()
caminhos = filedialog.askopenfilenames(
    title="Selecione até 35 planilhas preenchidas pelos professores",
    filetypes=[("Planilhas Excel", "*.xlsx")],
    multiple=True
)

# Limita a 35 arquivos
arquivos = list(caminhos)[:35]
print(f"{len(arquivos)} arquivo(s) selecionado(s).")

# Cria dicionário com dados por turma
dados_por_turma = {turma: [] for turma in abas_turmas}

# Lê e coleta os dados
for arquivo in arquivos:
    for turma in abas_turmas:
        try:
            df = pd.read_excel(arquivo, sheet_name=turma)
            df['TURMA'] = turma
            df['ARQUIVO'] = arquivo.split('/')[-1]
            dados_por_turma[turma].append(df)
        except Exception:
            print(f"Aba '{turma}' não encontrada em '{arquivo}'.")

# Consolida e salva em arquivos por turma
for turma, lista in dados_por_turma.items():
    if lista:
        resultado = pd.concat(lista, ignore_index=True)
        resultado.to_excel(f"{turma}_consolidado.xlsx", index=False)

print("Consolidação finalizada.")
'''

with open("consolida_planilhas.py", "w") as f:
    f.write(codigo)

from google.colab import files
files.download("consolida_planilhas.py")
```
